## Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('/content/drive/MyDrive/Employee.csv')
data_set = pd.DataFrame(data)
```

# Taking a quick look at the data

```python
data_set.head(10)
```

We can use the function `tail()` to get the last rows of the dataset

```python
data_set.tail(10)
```

With the describe function to get a brief statistical summary of the data

```python
data_set.describe(include=['O'])
```

# Getting unique values from a column


We can use the unique function to get all the unique values

```python
levels_of_education = data_set['Education'].unique()
print(levels_of_education)
```

Transposing the column to get a better view of the data and  the value_counts function counts the occurrences of each unique value.

```python
pd.DataFrame(data_set['Education'].value_counts()).T
```

```python
pd.DataFrame(data_set[['Education','Gender']].value_counts()).T
```

Selecting multiple specific colmuns to display we can use this method below:

 `dataset[[column1,column2]]`

```python
pd.set_option('display.max_rows', 100)
data_set[['Education','Age','City']]
```

```python
critical_vals = pd.DataFrame(data_set[['Gender','PaymentTier']].value_counts()).T
critical_vals
```

# Selecting rows with boolean indexing
  Boolean indexing allows you to select rows based on a condition. In the example below we are selecting rows of emplotees who have more that 2 years of experience.

```python
data_set[data_set['ExperienceInCurrentDomain']>3]
```

```python
data_set[data_set['Age']>27]
```

# Grouping values with pandas

Learning how to use the groupby function to group the employees by education. In this example below we are grouping the dataframe by education, we are also using the value_counts function to see how many males and females in each level of education

```python
grouped_by_education =  pd.DataFrame(data_set.groupby('Education')['Gender'].value_counts())
grouped_by_education
```

```python
 pd.set_option('display.max_rows', 200)
 pd.DataFrame(data_set.groupby('PaymentTier')[['Gender','Education']].value_counts())
```
